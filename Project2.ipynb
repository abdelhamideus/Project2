# 3. Data Wrangling:

```python

df.isna().sum()
```

```python
df['Goal_assist'] = df['Goal_assist'].fillna('No Assist')
```

```python
df
```

```python
df['Goal_assist'] = df['Goal_assist'].str.replace('\t','No Assist',regex=False)
df
```

```python
df['Playing_Position'] = df['Playing_Position'].ffill()
df['Type'] = df['Type'].ffill()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Minute'] =df['Minute'].astype(str).str.split('+').str[0]
df['Minute'] =df['Minute'].astype(int)
```

```python
df['Minute']
```

```python
df.head()
```

```python
df['At_score'] =df['At_score'].astype(str).str.split(':').str[0]
df['At_score'] =df['At_score'].astype(int)
```

```python
df['At_score']
```

```python
df.head()
```

```python
df.drop('At_score',axis=1)
```

```python
df.drop('Result',axis=1)
```

```python
df.drop('Matchday',axis=1)
```

```python
df = df.rename(columns={'Assist_Maker':'Assister','Playing_Position':'Position','Goal_assist':'Assister'})
df.drop(['Result','At_score','Matchday'],axis=1)
```

```python
df['Type'].unique()
```

```python
df['Position']=df['Position'].replace('LW ','LW')
df['Position']=df['Position'].replace('CF ','CF')
df['Position'].unique()
```

```python
df['Season'] = df['Season'].replace('Dec-13','12/13')
```

# 4. Feauture Engineering:

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day_name()
```

```python
df
```

```python
df['Match_ID'] = df['Date'].astype(str) + ',' + df['Club'] + ',' + df['Opponent']

df
```

# 5. Exploratory Data Analysis (EDA) & Visualization:

```python
most_goals_in_a_single_match = df['Match_ID'].value_counts().reset_index().head(10)
most_goals_in_a_single_match
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=most_goals_in_a_single_match, 
            x='Match_ID', 
            y='count',
            hue='count',
            palette='viridis')

plt.title("Iconic Matches", fontsize=16, fontweight='bold')
plt.xlabel("Matches")
plt.ylabel("Goals Scored")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('Iconic_Matches.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
fav_opponents = df['Opponent'].value_counts().reset_index().head(10)
fav_opponents
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=fav_opponents, 
            x='Opponent', 
            y='count',
            hue='count',
            palette='viridis')

plt.title("Favourite Opponents", fontsize=16, fontweight='bold')
plt.xlabel("Opponents")
plt.ylabel("Goals Scored")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('Favourite_Opponents.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
